##Problema 1
Entre todos os retângulos possíveis, de area igual a 100. Encontre aquele com o menor perímetro.

Para encontrar o retângulo com o menor perímetro e área igual a 100, podemos começar observando as possíveis dimensões do retângulo. A área de um retângulo é dada por $A = \text{comprimento} \times \text{largura}$, e o perímetro é dado por $P = 2 \times (\text{comprimento} + \text{largura})$.

Dado que a área é 100, podemos expressar uma das dimensões em termos da outra. Assumindo que o comprimento seja $L$ e a largura seja $W$, temos:

$$A = L \times W = 100$$

Podemos então reescrever $L$ em termos de $W$:

$$L = \frac{100}{W}$$

Substituindo $L$ na fórmula do perímetro, obtemos:

$$P = 2 \times \left(\frac{100}{W} + W\right)$$

Derivando $P$ em relação a $W$ e igualando a zero, temos:

$$\frac{dP}{dW} = 2 \times \left(-\frac{100}{W^2} + 1\right) = 0$$

Isolando $W$, obtemos:

$$-\frac{100}{W^2} + 1 = 0$$

$$-\frac{100}{W^2} = -1$$

$$W^2 = 100$$

$$W = \sqrt{100}$$

$$W = 10$$

Portanto, a largura do retângulo com o menor perímetro e área igual a 100 é 10. Substituindo $W$ de volta na expressão para $L$, temos:

$$L = \frac{100}{10} = 10$$

Assim, o retângulo com o menor perímetro e área igual a 100 tem dimensões de 10 unidades de largura por 10 unidades de comprimento.

```python
import numpy as np

area = 100
min_perimeter = float('inf')
best_rectangle = None

for length in range(1, area + 1):
    width, remainder = divmod(area, length)
    if remainder == 0:
        perimeter = 2 * (length + width)
        if perimeter < min_perimeter:
            min_perimeter = perimeter
            best_rectangle = (length, width)

best_rectangle
```

##Problema 2
Você foi encarregado de determinar o melhor preço e o número de itens que devem ser vendidos para maximizar o lucro da sua empresa. Uma pesquisa indicou que se você fixar o preço em $1,50$, será capaz de vender 5000 itens. Contudo, a cada $0.10$ de redução no preço de venda você terá um aumento de 1000, no múmero de itens vendidos. Considere que seus custos fixos são de $2000.00$ e que existe um custo marginal de $0.50$ por item produzido. Encontre o preço de venda que gerará o maior lucro.

Para encontrar o preço de venda que gerará o maior lucro, precisamos primeiro calcular o lucro para cada preço e, em seguida, encontrar o preço que resulta no maior lucro. Vamos fazer isso passo a passo:

\begin{align*}
\text{Lucro total} & = \text{Receita total} - \text{Custo total} \\
\text{Receita total} & = \text{Preço de venda} \times \text{Número de itens vendidos} \\
\text{Custo total} & = \text{Custos fixos} + (\text{Custo marginal} \times \text{Número de itens vendidos})
\end{align*}

Agora, vamos calcular o lucro para cada preço:

Para $p = 1.50$:

\begin{align*}
\text{Número de itens vendidos}: & 5000 \\
\text{Receita total}: & 1.50 \times 5000 = 7500.00 \\
\text{Custo total}: & 2000.00 + (0.50 \times 5000) = 4500.00 \\
\text{Lucro total}: & 7500.00 - 4500.00 = 3000.00 \\
\end{align*}

Para $p = 1.40$:

\begin{align*}
\text{Número de itens vendidos}: & 5000 + 1000 = 6000 \\
\text{Receita total}: & 1.40 \times 6000 = 8400.00 \\
\text{Custo total}: & 2000.00 + (0.50 \times 6000) = 5000.00 \\
\text{Lucro total}: & 8400.00 - 5000.00 = 3400.00 \\
\end{align*}

E assim por diante. Podemos calcular para mais alguns preços e encontrar o que gera o maior lucro. Vamos tentar para mais alguns preços e encontrar o preço que maximiza o lucro.

```python
import numpy as np

# Custos fixos
custos_fixos = 2000.00

# Custo marginal por item
custo_marginal = 0.50

# Preço inicial e número inicial de itens vendidos
preco = 1.50
num_itens_vendidos = 5000

# Inicialização do lucro máximo e do preço que gera o máximo lucro
max_lucro = 0
preco_max_lucro = 0

# Loop para testar diferentes preços
while preco >= 0.50:
    # Cálculo do lucro
    receita = preco * num_itens_vendidos
    custo_total = custos_fixos + custo_marginal * num_itens_vendidos
    lucro = receita - custo_total

    # Atualização do preço que gera o máximo lucro
    if lucro > max_lucro:
        max_lucro = lucro
        preco_max_lucro = preco

    # Redução do preço e aumento no número de itens vendidos
    preco -= 0.10
    num_itens_vendidos += 1000

print(f"O preço que gera o maior lucro é {preco_max_lucro:.2f} e o lucro máximo é {max_lucro:.2f}")
```

2. Raízes da função: Encontrar os valores de $x$ onde $f(x) = 0$ nos dará as raízes da função.
3. Intervalos de pontos críticos: Para determinar os intervalos onde existem pontos críticos, vamos calcular a primeira e segunda derivadas da função e analisar seus sinais.

```python
import sympy as sp

# Definindo a variável simbólica x
x = sp.symbols('x')

# Definindo a função
f = x**4 - 6*x**2

# Encontrando as raízes da função (onde f(x) = 0)
roots = sp.solve(f, x)

# Calculando as derivadas da função
f_prime = sp.diff(f, x)
f_double_prime = sp.diff(f_prime, x)

# Encontrando os pontos críticos (onde a primeira derivada é zero ou indefinida)
critical_points = sp.solve(f_prime, x)

# Imprimindo as raízes da função e os pontos críticos
print("Raízes da função (f(x) = 0):", roots)
print("Pontos Críticos (f'(x) = 0):", critical_points)
```

b) Agora, para encontrar os pontos críticos, máximo local, mínimo local ou sela (caso existam), e traçar os gráficos da função, da primeira derivada e da segunda derivada, marcando os pontos críticos, podemos seguir estes passos:

1. Pontos críticos: Encontrar os valores de $x$ onde a primeira derivada é zero ou indefinida.
2. Determinar a natureza dos pontos críticos: Usar a segunda derivada para determinar se os pontos críticos são máximos, mínimos ou pontos de sela.
3. Gráficos das derivadas: Traçar os gráficos da primeira e segunda derivadas e marcar os pontos críticos.


```python
import sympy as sp

# Definindo a variável simbólica x
x = sp.symbols('x')

# Definindo a função
f = x**4 - 6*x**2

# Calculando as derivadas
f_prime = sp.diff(f, x)
f_double_prime = sp.diff(f_prime, x)

# Encontrando os pontos críticos (zeros da primeira derivada)
critical_points = sp.solve(f_prime, x)

# Encontrando os pontos onde a segunda derivada se anula (possíveis pontos de sela)
saddle_points = sp.solve(f_double_prime, x)

# Gerando valores para x
x_vals = np.linspace(-3, 3, 1000)
y = np.array([f.subs(x, val) for val in x_vals])
y_prime = np.array([f_prime.subs(x, val) for val in x_vals])
y_double_prime = np.array([f_double_prime.subs(x, val) for val in x_vals])

# Plotando o gráfico da função, da primeira derivada e da segunda derivada
plt.figure(figsize=(18, 10))
plt.subplot(311)
plt.plot(x_vals, y, label='f(x)')
plt.scatter(critical_points, [f.subs(x, cp) for cp in critical_points], color='red', label='Pontos Críticos')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da função $f(x)$ com pontos críticos')
plt.grid(True)
plt.legend()

plt.subplot(312)
plt.plot(x_vals, y_prime, label="f'(x)")
plt.scatter(critical_points, [0]*len(critical_points), color='red', label='Pontos Críticos')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.title("Gráfico da primeira derivada $f'(x)$ com pontos críticos")
plt.grid(True)
plt.legend()

plt.subplot(313)
plt.plot(x_vals, y_double_prime, label="f''(x)")
plt.scatter(saddle_points, [0]*len(saddle_points), color='blue', label='Pontos de Sela')
plt.xlabel('x')
plt.ylabel("f''(x)")
plt.title("Gráfico da segunda derivada $f''(x)$ com pontos de sela")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

print("Pontos Críticos:", critical_points)
print("Pontos de Sela:", saddle_points)
```
