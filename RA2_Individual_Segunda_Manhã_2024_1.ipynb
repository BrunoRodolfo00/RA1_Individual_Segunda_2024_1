**Alternativa A**
1. Encontrar a derivada da função \( f(x) = x^4 - 4x^3 + 4x^2 \):

$$ f'(x) = \frac{d}{dx}(x^4) - \frac{d}{dx}(4x^3) + \frac{d}{dx}(4x^2) $$

$$ f'(x) = 4x^3 - 12x^2 + 8x $$

2. Igualar a derivada a zero para encontrar os pontos críticos:

$$ 4x^3 - 12x^2 + 8x = 0 $$

3. Fatorar \( 4x \) da equação:

$$ 4x(x^2 - 3x + 2) = 0 $$

4. Fatorar \( x^2 - 3x + 2 \):

$$ 4x(x - 1)(x - 2) = 0 $$

Portanto, os pontos críticos são \( x = 0 \), \( x = 1 \) e \( x = 2 \).

5. Encontrar a segunda derivada da função \( f(x) \):

$$ f''(x) = \frac{d}{dx}(4x^3 - 12x^2 + 8x) $$

$$ f''(x) = 12x^2 - 24x + 8 $$

6. Avaliar a segunda derivada nos pontos críticos:

- Para \( x = 0 \):
$$ f''(0) = 8 $$, positivo, então \( x = 0 \) é um mínimo local.

- Para \( x = 1 \):
$$ f''(1) = -4 $$, negativo, então \( x = 1 \) é um máximo local.

- Para \( x = 2 \):
$$ f''(2) = 8 $$, positivo, então \( x = 2 \) é um mínimo local.

Portanto, os pontos críticos são:
- \( x = 0 \) é um mínimo local,
- \( x = 1 \) é um máximo local, e
- \( x = 2 \) é um mínimo local.

$$

Encontramos os pontos críticos igualando a derivada primeira a zero:
$$
0 = 2e^{2x} - e^x
$$

Isso ocorre quando $$ e^x = 0 $$ (o que não é possível) ou quando $$ 2e^x - 1 = 0 $$. Resolvendo $$ 2e^x - 1 = 0 $$, obtemos $$ x = \ln\left(\frac{1}{2}\right) $$.

3. **Análise dos pontos críticos:**
Para determinar a natureza dos pontos críticos, usamos a segunda derivada:
$$
f''(x) = \frac{d^2}{dx^2} (2e^{2x} - e^x) = 8e^{2x} - e^x
$$

Avaliando a segunda derivada em $$( x = \ln\left(\frac{1}{2}\right) )$$, obtemos $$( f''\left(\ln\left(\frac{1}{2}\right)\right) = \frac{3}{2} )$$, que é positiva. Portanto, $$ x = \ln\left(\frac{1}{2}\right) $$ é um ponto de mínimo local da função $$ f(x) $$.

Encarando a vermelhidão do por do sol e sentindo o peso dos anos, um velho pescador precisa atravessar um rio para voltar a sua casa o mais rápido possível. Complicada será esta jornada já que sua casa está na margem oposta de um rio de águas profundas, negras e caudalosas com $10m$ entre margens. Os anos de experiência fazem com que ele tenha certeza que consegue remar a $2m/s$ e caminhar a $4m/s$ e que a correnteza do rio não o deixará cruzá-lo perpendicularmente. Ainda assim, ele precisa encontrar o tempo exato de remadas e caminhadas para chegar a usa casa no menor tempo possível.

Não é uma decisão fácil, nesta idade avançada, o pescador não pode caminhar e remar mais de uma vez e não será capaz de tomar a decisão correta sem a sua ajuda. Então observe que ele está na margem do rio e deve escolher o quanto deve caminhar, ou remar, para encontrar o menor tempo $T(x)$.

A figura a seguir representa um diagrama da situação real do pescador e você pode usar métodos algébricos ou computacionais para resolver a questão.

**Rubrica de Avaliação**: até dois pontos se o mínimo for corretamente encontrado.

```python
def tempo_total(x, v):
    # Tempo para caminhar até o ponto de partida da travessia
    tempo_caminhada = x / 4
    # Tempo para remar até a margem oposta
    tempo_remar = (10 - x) / (2 + v)
    # Tempo total é a soma dos tempos de caminhada e remo
    return tempo_caminhada + tempo_remar

# Encontrar a velocidade da correnteza do rio que minimiza o tempo total
v = 0
min_tempo = float('inf')
for i in range(1, 11):
    tempo = tempo_total(i, 2)
    if tempo < min_tempo:
        min_tempo = tempo
        v = i

# Calcular a distância que o pescador deve caminhar para minimizar o tempo total
x = (10 + v) / 2

print(f"Para minimizar o tempo total, o pescador deve caminhar {x} metros antes de começar a remar.")
print(f"O tempo total mínimo para atravessar o rio é de {min_tempo} segundos.")
```

Seja a função $f(x, y) = x^2y + 2x + y^3$. Encontra as derivadas faciais.

**Rubrica de Avaliação**: até um ponto se as derivadas forem corretamente encontradas.



```python
from sympy import symbols, diff

# Definindo as variáveis
x, y = symbols('x y')

# Definindo a função
f = x**2 * y + 2*x + y**3

# Calculando as derivadas parciais
df_dx = diff(f, x)
df_dy = diff(f, y)

# Exibindo as derivadas parciais
print("Derivada parcial em relação a x:", df_dx)
print("Derivada parcial em relação a y:", df_dy)
```
